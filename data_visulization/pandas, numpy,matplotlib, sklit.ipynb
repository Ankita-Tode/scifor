```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import matplotlib
%matplotlib inline
```

```python
population_df=pd.read_csv('world_population.csv')
```

```python
sns.set_style('darkgrid')
matplotlib.rcParams['font.size']=14
matplotlib.rcParams['figure.figsize']=(9,5)
matplotlib.rcParams['figure.facecolor']='#00000004'
```

```python
highest_population=population_df.sort_values('2020 Population',ascending=False).head(10)
```

```python
highest_population
```

```python
plt.figure(figsize=(20,10))
plt.title("plot showing countries and their population")
sns.barplot(x='Country/Territory',y='2020 Population',data=highest_population)
```

```python
highest_populaion_density=population_df.sort_values('Density (per km²)',ascending=False).head(10)
```

```python
highest_populaion_density
```

```python
plt.figure(figsize=(20,10))
plt.title("plot showing the world's most densely populated countries")

sns.barplot(x='Country/Territory',y='Density (per km²)',data=highest_populaion_density)
```

```python
highest_populaion_growth=population_df.sort_values('Growth Rate',ascending=False).head(10)
highest_populaion_growth
```

```python
plt.figure(figsize=(20,10))
plt.title("plot showing Growth Rate Of Countries")

sns.barplot(x='Country/Territory',y='Growth Rate',data=highest_populaion_growth)
```
